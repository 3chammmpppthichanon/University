```python
import cv2
import numpy as np
from utils import plt_image
```

```python
img = cv2.imread('./images/shade.png', cv2.IMREAD_COLOR)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt_image(img_rgb)
```

```python
img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
plt_image(img_hsv)
```

```python
h, s, v = cv2.split(img_hsv)
```

```python
h_2 = h.astype(np.uint16)* 2
mask_h = ((h_2 >= 160) & (h_2 <= 270))
s[mask_h] = s[mask_h]* 0.5 
new_hsv = cv2.merge([h, s, v])

new_hsv_rgb = cv2.cvtColor(new_hsv, cv2.COLOR_HSV2RGB)
plt_image(new_hsv_rgb)
```
