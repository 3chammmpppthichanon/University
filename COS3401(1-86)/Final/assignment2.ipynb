```python
import cv2
import numpy as np
from utility import *
```

```python
img = cv2.imread('./images/body_color.jpg', cv2.IMREAD_COLOR_RGB)
```

```python
show_image_plt(img)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
h, _, _ = cv2.split(hsv)
```

```python
h_double = h.astype(np.uint16) * 2
mask_h = ((h_double >= 1) & (h_double <= 35))
result = (mask_h.astype(np.uint8)) * 255
```

```python
show_image_cv2(result)
```

```python
show_image_cv2(result)
```

```python
result_blur = cv2.medianBlur(result, 5)
```

```python
show_image_cv2(result_blur)
```
