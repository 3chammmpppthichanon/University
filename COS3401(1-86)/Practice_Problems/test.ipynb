```python
import cv2
import numpy as np
from utility import *
```

```python
img = cv2.imread('../Example/images/images/shade.png', cv2.IMREAD_COLOR)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
show_image_plt(img_rgb)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
show_image_plt(img_hsv)
```

```python
h, s, v = cv2.split(img_hsv)
```

```python
h_2 = h.astype(np.float16) * 2
mask_h = ((h_2 >= 160) & (h_2 <= 270))
s[mask_h] = s[mask_h] * 0.5

new_img_hsv = cv2.merge([h,s,v])

new_hsv_rgb = cv2.cvtColor(new_img_hsv, cv2.COLOR_HSV2RGB)
show_image_plt(new_hsv_rgb)
```
