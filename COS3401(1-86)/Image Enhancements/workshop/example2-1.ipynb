```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_cv(img):
    cv2.imshow("xxxx", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def show_plt(img, opt=0):
    if opt:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
x =  cv2.imread('../images/mdb.bmp', cv2.IMREAD_GRAYSCALE)
print(type(x), x.shape)
```

```python
show_cv(x)
```

```python
show_plt(x, 1)
```

## Image Inversion

```python
y = 255 - x
```

```python
show_plt(y, 1)
```

```python
x = range(7)
print(type(x), len(x))
```

```python
m,n = x.shape
o = np.zeros_like(x, dtype=np.uint8)
for i in range(m):
    for j in range(n):
        o[i,j] = 255 - x[i,j]

show_plt(o, 1)
```
