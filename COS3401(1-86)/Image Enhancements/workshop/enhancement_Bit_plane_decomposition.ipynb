```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bit_plane_slicing(img, bit):
    bit_img = img.copy() # Create a copy of the image to avoid modifying the original
    bit_img = np.right_shift(bit_img, bit) # Shift the bits to the right by 'bit' positions
    bit_img = np.bitwise_and(bit_img, 1) 
    bit_img = (bit_img * 255).astype(np.uint8)
    return bit_img
```

```python
# Reading a image
img = cv2.imread('../../images/mandril_color.png', 0)
print(img.shape)
```

```python
planes = []
planes.append(img)
for i in range(8):
    planes.append(bit_plane_slicing(img, i))
```

```python
plt.figure(figsize=(10,10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(planes[i], cmap='gray')
    plt.title(f'Bit plane {i-1}')
    plt.xticks([])
    plt.yticks([])
plt.show()
```
