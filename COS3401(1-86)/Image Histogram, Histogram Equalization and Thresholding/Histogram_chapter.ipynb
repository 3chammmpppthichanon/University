## <span style='color:blue;'>What is a image histogram?</span>  
A histogram in image processing is a graphical representation of the distribution of pixel intensities in a digital image. It quantifies the number of pixels for each intensity value considered. 




Here's a more detailed explanation:

 **<span style='color:lightblue;'>Components of an Image Histogram</span>**

1. **<span style='color:lightgreen;'>Axes :</span>**
    - The X-axis represents the intensity values (usually ranging from 0 to 255 in an 8-bit image).
    - The Y-axis represents the number of pixels in the image that have each intensity value.

2. **<span style='color:lightgreen;'>Intensity :</span>**
    - In a grayscale image, intensity values range from 0 (black) to 255 (white).
    - In a color image, histograms can be created for each color channel (Red, Green, Blue).

3. **<span style='color:lightgreen;'>Bin :</span>**
    - Each column in the histogram represents a 'bin' that counts how many pixels in the image have a specific intensity value.

**<span style='color:Tomato;'>Uses of Histograms in Image Processing (in this chapter)</span>**

1. **<span style='color:orange;'>Image Enhancement :</span>**
    - Histograms are used in techniques like histogram equalization, which improves the contrast in an image.



2. **<span style='color:orange;'>Thresholding (Background and Foreground Separation) :</span>**
    - Histograms can help in determining a good threshold value to separate objects in an image.
    - It helps in distinguishing between different components of the image based on their pixel intensities.

       


#### <span style='color:lightblue;'>Histogram Processing in Python</span>  
1. **<span style='color:lightgreen;'>Import Libraries</span>**
    - Import the necessary libraries.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

2. **<span style='color:lightgreen;'>Read an Image</span>**
    - Load an image using OpenCV.

```python
img = cv2.imread('./images/bank3.jpg', cv2.IMREAD_COLOR)
print(img.shape)
```

```python
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.show()
```

3. **<span style='color:lightgreen;'>Calculate Histogram using cv2 library</span>**
    - Compute the histogram of the image.

```python
hist_0 = cv2.calcHist([gray], [0], None, [256], [0, 256])
```

```python
print(type(hist_0), hist_0.shape)
```

4. **<span style='color:lightgreen;'>Plot Histogram</span>**
    - Use Matplotlib to display the histogram.

```python
plt.figure(figsize=(4,2))
plt.plot(hist_0)
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
# plt.savefig('histogram.png')
```

5. **<span style='color:lightgreen;'>Calculate Histogram without using cv2 library</span>**
    - Here's a sample Python script to create a histogram of an image without using a histogram-specific library:

```python
hist_1 = np.zeros(256)
row, col = gray.shape
for i in range(row):
    for j in range(col):
        intens = img[i, j]
        hist_1[intens] += 1
```

```python
plt.figure(figsize=(4,2))
plt.plot(hist_1)
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
