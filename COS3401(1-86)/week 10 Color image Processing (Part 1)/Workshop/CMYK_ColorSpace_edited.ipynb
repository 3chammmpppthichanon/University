```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showImage(img, gray):
    plt.axis('off')
    if gray:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(img)
    
    plt.show()
```

```python
img = cv2.imread('../images/images/bank.png')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
showImage(img, False)
```

```python
r, g, b = cv2.split(rgb)
showImage(r, False)
```
