```python
import cv2
import numpy as np

from utils import show_image
```

```python
img = cv2.imread('./images/body_color.jpg', cv2.IMREAD_COLOR_RGB)
```

```python
show_image(img, 'Origin (Color)')
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
h, _, _ = cv2.split(hsv)
```

```python
h_double = h.astype(np.uint16) * 2

mask_h = ((h_double >= 1) & (h_double <= 45))
result = (mask_h.astype(np.uint8)) * 255
```

```python
show_image(result, "H Color 1-45 Degree")
```

```python
result_blur = cv2.medianBlur(result, 5)
```

```python
show_image(result_blur, "Median Blur")
```

```python
gray = cv2.imread('./images/body_gray.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
show_image(gray, 'Origin (Gray)')
```

```python
gray[result_blur == 255] = gray[result_blur == 255] * 2
```

```python
show_image(gray, "Result")
```

```python
show_image(cv2.hconcat([result_blur, gray]), 'result')
```

```python
cv2.imwrite('./outs/assign_2-1.png', result_blur.astype(np.uint8))
cv2.imwrite('./outs/assign_2-2.png', gray.astype(np.uint8))
```
