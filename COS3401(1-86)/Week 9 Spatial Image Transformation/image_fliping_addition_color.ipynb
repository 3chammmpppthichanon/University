### <span style='color:lightgreen;'>Simple Image Manipulations</span>

**<span style='color:lightgreen;'>Horizontally and Vertically Flipped Image</span>** Flipping an image horizontally means mirroring it along its vertical axis. Flipping an image vertically means mirroring it along its horizontal axis.
1. **<span style='color:lightgreen;'>Mathematical Equation of horizontalflipping:</span>** Let's say we have an image represented by a matrix I, where I(x, y) represents the pixel value at coordinates (x, y). If(x,y)isapixellocation,thenitsnewlocationafterahorizontalflipis(x,W1y)





2. **<span style='color:lightgreen;'>Mathematical Equation of verticallyflipping:</span>** A vertically flipped image is mirrored along the horizontal axis. The mathematical transformation is: If(x,y)isapixellocation,thenitsnewlocationafteraverticalflipis(H1x,y)





```python
def ShowImage(img):
    plt.axis("off")
    plt.imshow(img)
    plt.show()
```

```python
def flip_image(img, opt):
    r,c,n = img.shape
    out = np.zeros_like(img, dtype='uint8') 
    for i in range(r):
        for j in range(c):
            if opt == 0:
                #out[i,c-j-1] = img[i,j]
                out[i,j,:] = img[i,c-j-1,:]
            elif opt == 1:
                out[r-i-1,j,:] = img[i,j,:]
            elif opt == 2:
                out[r-i-1,c-j-1,:] = img[i,j,:]
            else:
                out[i,j,:] = img[i,j,:]
    return out.astype(np.uint8)
```

```python
img = cv2.imread("./images/lena_color_256.png")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
ShowImage(rgb)
print(rgb.shape)
```

```python
flip_img = flip_image(rgb, 2)
mul_img = cv2.hconcat([rgb, flip_img])
ShowImage(mul_img)
```

**<span style='color:lightgreen;'>Transposed Image</span>** Transposing an image means interchanging its rows and columns. It's like rotating the image by 90 degrees counter-clockwise and then flipping it horizontally.
1. **<span style='color:lightgreen;'>Mathematical Equation of diagonalflipping:</span>** A transposed image swaps the rows and columns, effectively flipping the image along the diagonal. The transformation for this is:  If<i>(x,y)</i>isapixellocation,thenitsnewlocationaftertranspositionis<i>(y,x)</i>




```python
def Transpost_image(img):
    r, c, n = img.shape
    out = np.zeros((c, r, n), dtype='uint8') 
    
    for i in range(c):
        for j in range(r):
            out[i, j,:] = img[j,i,:]

    return out.astype(np.uint8)
```

```python
Timg = Transpost_image(rgb)
ShowImage(Timg)
```
