# Import Modules

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Read an image

```python
img = cv2.imread(r'C:\Users\Thichanon\Desktop\University\COS3401\HomeWork\assignment_image\ivus.png', cv2.IMREAD_GRAYSCALE)
```

```python
cv2.imshow('Original image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

### Negative Transform

```python
negative_img = 255 - img
print(negative_img.shape, img.shape)
```

```python
cv2.imshow('Image negative transform', negative_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(negative_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure()
fig.add_subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.axis('off')

fig.add_subplot(1,2,2)
plt.imshow(negative_img, cmap='gray')
plt.axis('off')

plt.show()
```
