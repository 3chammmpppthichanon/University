# Import libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Read an image

```python
image = cv2.imread(r'C:\Users\Thichanon\Desktop\University\COS3401\Week6\images\bank3.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

## Cal hist

```python
hist_0 = cv2.calcHist([image], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(4,2))
plt.plot(hist_0)
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

5. **<span style='color:lightgreen;'>Calculate Histogram without using cv2 library</span>**
    - Here's a sample Python script to create a histogram of an image without using a histogram-specific library:

```python
hist_1 = np.zeros(256)
row, col = image.shape
for i in range(row):
    for j in range(col):
        intens = image[i, j]
        hist_1[intens] += 1
```

```python
plt.figure(figsize=(4,2))
plt.plot(hist_1)
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
