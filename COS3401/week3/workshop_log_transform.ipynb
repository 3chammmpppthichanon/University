## <span style='color:lightblue;'>Let's start</span>  

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### <span style='color:lightgreen;'>Read Image</span>  

```python
img = cv2.imread(r'C:\Users\Thichanon\Desktop\University\COS3401\images\dark2.png', 0)
```

```python
img_float32 = np.float32(img)
c = 255 / np.log(1 + np.max(img_float32))
print(np.amax(img_float32), np.log(1+ np.amax(img_float32)))
```

### <span style='color:lightgreen;'>Method I of The Logarithm Transformation</span>  

```python
log_img = c * np.log(1 + img_float32)
log_img = log_img.astype(np.uint8)
```

```python
cv2.imshow('result', log_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig = plt.figure()
fig.add_subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')

fig.add_subplot(1, 2, 2)
plt.imshow(log_img, cmap='gray')
plt.axis('off')

plt.show()
```

### <span style='color:lightgreen;'>Method II of  The Logarithm Transformation</span>  

```python
log_img1 = np.zeros_like(img, dtype=np.float32)
row, col = log_img1.shape
print(row, col)
```

```python
for i in range(row):
    for j in range(col):
        log_img1[i, j] = c * np.log(1 + img[i,j])

log_img1 = log_img1.astype(np.uint8)
```

```python
cv2.imshow('result1', log_img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig = plt.figure()
fig.add_subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.axis('off')

fig.add_subplot(1,2,2)
plt.imshow(log_img1, cmap='gray')
plt.axis('off')

plt.show()
```
