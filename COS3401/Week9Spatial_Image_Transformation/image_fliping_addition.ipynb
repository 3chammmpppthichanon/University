### <span style='color:lightgreen;'>Simple Image Manipulations</span>

**<span style='color:lightgreen;'>Horizontally and Vertically Flipped Image</span>** Flipping an image horizontally means mirroring it along its vertical axis. Flipping an image vertically means mirroring it along its horizontal axis.
1. **<span style='color:lightgreen;'>Mathematical Equation of horizontal flipping:</span>** Let's say we have an image represented by a matrix I, where I(x, y) represents the pixel value at coordinates (x, y). If (x,y) is a pixel location, then its new location after a horizontal flip is (x,W−1−y)
2. **<span style='color:lightgreen;'>Mathematical Equation of vertically flipping:</span>** A vertically flipped image is mirrored along the horizontal axis. The mathematical transformation is: If (x,y) is a pixel location, then its new location after a vertical flip is (H−1−x,y)




```python
def ShowImage(img):
    plt.axis("off")
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
def flip_image(img, opt):
    r,c = img.shape
    out = np.zeros_like(img, dtype='uint8') 
    for i in range(r):
        for j in range(c):
            if opt == 0:
                out[i,c-j-1] = img[i,j]
            elif opt == 1:
                out[r-i-1,j] = img[i,j]
                #out[i, j] = img[r-i-1, j]
            elif opt == 2:
                out[r-i-1,c-j-1] = img[i,j]
            else:
                out[i,j] = img[i,j]
    return out.astype(np.uint8)
```

```python
img = cv2.imread("./images/cameraman.png", 0)
```

```python
ShowImage(img)
print(img.shape)
```

```python
flip_img = flip_image(img, 2)
mul_img = cv2.hconcat([img, flip_img])
ShowImage(mul_img)
```

**<span style='color:lightgreen;'>Transposed Image</span>** Transposing an image means interchanging its rows and columns. It's like rotating the image by 90 degrees counter-clockwise and then flipping it horizontally.
1. **<span style='color:lightgreen;'>Mathematical Equation of diagonal flipping:</span>** A transposed image swaps the rows and columns, effectively flipping the image along the diagonal. The transformation for this is:  If <i>(x,y)</i> is a pixel location, then its new location after transposition is <i>(y,x)</i>




```python
def Transpost_image(img):
    r, c = img.shape
    out = np.zeros((c, r), dtype='uint8') 
    
    for i in range(c):
        out[i, :] = img[:, i]
    
    # for i in range(c):
    #     for j in range(r):
    #         out[i, j] = img[j,i]

    return out.astype(np.uint8)
```

```python
Timg = Transpost_image(img)
ShowImage(Timg)
```
