```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread(r'C:\Users\Thichanon\Desktop\University\COS3401\Week7-8\images\bank3.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(4,2))
plt.plot(hist)
plt.title('Histogram')
plt.xlabel('Pixel intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
thresh = 100
ret, bin_img = cv2.threshold(img, thresh, 255, cv2.THRESH_BINARY)

plt.imshow(bin_img, cmap='gray')
plt.axis('off')
plt.title('Thresholded Image')
plt.show()
```

```python
thresh = 150
bimg = np.zeros_like(img, dtype='uint8')
bimg[img > thresh] = 255
plt.imshow(bimg, cmap='gray')
plt.title('Thresholded Image')
plt.show()
```

```python
thresh = 150
out = np.zeros_like(img, dtype='uint8')
row, col = img.shape
for i in range(row):
    for j in range(col):
        if img[i,j] >= thresh:
            out[i,j] = 255
plt.imshow(out, cmap='gray')
plt.title('Thresholded Image')
plt.show()
```
