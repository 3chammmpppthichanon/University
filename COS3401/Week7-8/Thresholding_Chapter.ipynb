## <span style='color:lightblue;'>What is Image Thresholding?</span>  
Image thresholding is a simple, yet effective, technique in image processing and computer vision used to create binary images from grayscale or color images. The basic idea behind thresholding is to separate pixels into different groups based on their intensity values, typically to distinguish objects or features of interest in the image.

  

 **<span style='color:lightblue;'>Here's how it generally works:</span>**

1. **<span style='color:lightgreen;'>Selection of a Threshold Value:</span>**
    - A threshold value is chosen. This value is a specific intensity level which acts as a cutoff point.

2. **<span style='color:lightgreen;'>Pixel Classification:</span>**
    - Every pixel in the image is then compared to this threshold value. 
    - If a pixel's intensity is higher than the threshold, it is assigned one value (often white or '1'), and if it's lower, it is assigned another value (often black or '0').

3. **<span style='color:lightgreen;'>Binary Image Creation: </span>**
    -  As a result of this process, the original image is converted into a binary image, consisting of only two pixel values, representing the foreground and the background. T

**<span style='color:Tomato;'>There are different methods to perform thresholding:</span>**

1. **<span style='color:orange;'>Global Thresholding: </span>**
    - A single threshold value is used for the entire image. This method is effective when there's a high contrast between the object of interest and the background.



2. **<span style='color:orange;'>Adaptive Thresholding:</span>**
    - Different threshold values (Local Thresholding) are used for different regions of the image. This method is useful when lighting conditions vary across different parts of the image.



3. **<span style='color:orange;'>Intermean and Otsus Methods: </span>**
    - An automatic way to select the threshold value by maximizing the variance between classes of pixels.

**<span style='color:Tomato;'>Thresholding:</span>** is widely used in many areas such as document scanning and processing, medical imaging, machine vision, and face recognition. It's a fundamental step in many image processing pipelines and is valued for its simplicity and efficiency in segmenting images into meaningful parts.

#### <span style='color:lightblue;'>Image Thresholding with pre-defined value</span>  
1. **<span style='color:lightgreen;'>Import Libraries</span>**
    - Import the necessary libraries.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

2. **<span style='color:lightgreen;'>Read an Image</span>**
    - Load an image using OpenCV.

```python
img = cv2.imread('./images/bank3.jpg', )
```

```python
plt.imshow(img, cmap='gray')
plt.axis("off")
plt.show()
```

3. **<span style='color:lightgreen;'>Calculate Histogram using cv2 library</span>**
    - Compute the histogram of the image.

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
```

4. **<span style='color:lightgreen;'>Plot Histogram</span>**
    - Use Matplotlib to display the histogram.

```python
plt.figure(figsize=(4,2))
plt.plot(hist)
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
# plt.savefig('histogram.png')
```

5. **<span style='color:lightgreen;'>Apply Threshold with fix value via cv2 library</span>**
    - Choose a threshold value based on the histogram and apply thresholding.

```python
thresh = 100

ret, bin_img = cv2.threshold(img, thresh, 255, cv2.THRESH_BINARY)

plt.imshow(bin_img, cmap='gray')
plt.title('Thresholded Image')
plt.show()
```

6. **<span style='color:lightgreen;'>Apply Threshold with fix value</span>**
    - Choose a threshold value based on the histogram and apply thresholding.

```python
thresh = 150
bimg = np.zeros_like(img, dtype='uint8')
bimg[img>=thresh] = 255
plt.imshow(bimg, cmap='gray')
plt.title('Thresholded Image')
plt.show()
```

```python
thresh = 150
out = np.zeros_like(img, dtype='uint8')
row, col = img.shape
for i in range(row):
    for j in range(col):
        if img[i,j] >= thresh:
            out[i,j] = 255
plt.imshow(out, cmap='gray')
plt.title('Thresholded Image')
plt.show()
```
