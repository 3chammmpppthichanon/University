## <span style='color:lightgreen;'>GREAY LEVEL IMAGE</span> 
A **<span style='color:lightblue;'>digital gray level image</span>** is a type of digital image where each pixel represents a single intensity value, which corresponds to the amount of light or brightness at that point. These images are often referred to as grayscale images and are composed exclusively of shades of gray, ranging from black at the weakest intensity to white at the strongest intensity.




### <span style='color:lightgreen;'>Key Characteristics of Digital Gray Level Images</span> 

1. **Pixel Intensity**: Each pixel in a gray level image has an intensity value that indicates its brightness. This value is typically represented as an integer within a specific range, such as 0 to 255 for 8-bit images, where 0 represents black and 255 represents white.



2. **Bit Depth**: The number of bits used to represent each pixel determines the number of different gray levels available. Common bit depths include:
   - **8-bit**: 256 gray levels (0-255)
   - **16-bit**: 65,536 gray levels (0-65,535)

3. **Quantization**: The process of converting the continuous range of intensity values into discrete levels. This is essential for digital representation and storage of images.

4. **Applications**: Gray level images are widely used in various fields such as medical imaging, remote sensing, and digital photography due to their simplicity and effectiveness in representing intensity variations without the complexity of color information.



###  <span style='color:lightgreen;'>Comparison with Color Images</span> 

| Feature              | Gray Level Image                    | Color Image                          |
|----------------------|-------------------------------------|--------------------------------------|
| Pixel Representation | Single intensity value              | Multiple values (e.g., RGB channels) |
| Storage Requirement  | Lower (e.g., 8 bits per pixel)      | Higher (e.g., 24 bits per pixel)     |
| Complexity           | Simpler                             | More complex                         |
| Use Cases            | Medical imaging, document scanning  | Photography, video, graphic design   |

### <span style='color:lightblue;'>Examples of Use</span>  

- **<span style='color:orange;'>Medical Imaging:</span>** High bit-depth gray level images (e.g., 12-bit or 16-bit) are used to capture fine differences in tissue density in X-ray or MRI scans.



- **<span style='color:orange;'>Document Scanning:</span>** Gray level images are used to digitize text documents, where color information is not necessary.



- **<span style='color:orange;'>Remote Sensing:</span>** Satellite images often use gray level representation to highlight variations in terrain or vegetation.



In summary, a digital gray level image is a fundamental type of image in digital imaging that captures and represents the intensity of light at each pixel, making it a versatile tool in various scientific and practical applications.

#### <span style='color:yellow;'>Method III : Linear Equation</span>   
**<span style='color:orange;'>Coordinate Pixels Visiting</span>**

```python
fp = "./images/mandril_color.tif"
img = cv2.imread(fp)
```

```python
r = img[:,:,2].astype(np.float32)
g = img[:,:,1].astype(np.float32)
b = img[:,:,0].astype(np.float32)
# b,g,r = cv2.split(img)
m,n = r.shape
```

```python
#gray = np.zeros((m,n), dtype='float')
gray = np.zeros_like(r, dtype='float')

for i in range(m):
    for j in range(n):
        gray[i,j] = 0.2989*r[i,j] + 0.5870*g[i,j] + 0.1140*b[i,j]
```

```python
plt.axis("off")
plt.imshow(gray, cmap='gray')
plt.show()
cv2.imwrite('./outs/output.png',gray.astype(np.uint8))
```

## <span style='color:lightgreen;'>What is the digital image histogram?</span> 
A digital image histogram is a graphical representation of the tonal distribution in a digital image. Here are the key points about digital image histograms:
1. **<span style='color:orange;'>Definition:</span>**   An image histogram plots the number of pixels for each tonal value in an image.
2. **<span style='color:orange;'>Visual representation:</span>** The horizontal axis represents the tonal variations from dark to light (left to right), while the vertical axis shows the number of pixels at each tonal level.
3. **<span style='color:orange;'>Tonal range:</span>** 
   - Left side: Represents shadows and dark areas
   - Middle: Represents midtones
   - Right side: Represents highlights and bright areas[8][14]
4. **<span style='color:orange;'>Information provided:</span>**
   - Shows the distribution of pixel intensities in an image[7]
   - Helps assess exposure, contrast, and dynamic range[3][8]



#### <span style='color:yellow;'>Method I: Create Histogram array using cv2.calcHist </span>      

```python
fp = "./images/mandril_color.tif"
img = cv2.imread(fp)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
i = 2
hist= cv2.calcHist([img], [i], None, [256], [0,256])

fig = plt.figure()
fig.add_subplot(1, 2, 1)
plt.imshow(img[:,:,i], cmap = 'gray')
plt.axis('off')
plt.title("grayscale image")
  
fig.add_subplot(1, 2, 2)
plt.plot(hist, color='b')
plt.title('Histogram for gray scale image')
plt.show()
```

#### <span style='color:yellow;'>Method III: Show Histogram using frequency array</span>        

```python
fp = "./images/lena_color_256.tif"
img = cv2.imread(fp)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

hist_img = np.zeros(256)
row, col = gray.shape
for i in range(row):
    for j in range(col):
        intens = gray[i, j]
        hist_img[intens] += 1

plt.figure(figsize=(4,2))
plt.plot(hist_img)
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

### <span style='color:lightgreen;'>Histogram of Color Image </span>
